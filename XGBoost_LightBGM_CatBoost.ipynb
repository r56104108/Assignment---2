```python
import pandas as pd
import numpy as np
```

```python
%cd /content/drive/My Drive/Assignment-2/
df = pd.read_csv("train.csv")
df = df.drop([df.columns[0]], axis = 1)
new_raw_data = pd.get_dummies(df)
new_raw_data
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(new_raw_data, test_size=.2, random_state=41)

X_train = train.drop(columns=['is_claim'])
y_train = train["is_claim"]
X_test = test.drop(columns=['is_claim'])
y_test = test["is_claim"]
```

# XGBoost

```python
from xgboost import XGBClassifier

xgboostModel = XGBClassifier(n_estimators=100)
xgboostModel.fit(X_train, y_train)
print(xgboostModel.score(X_test,y_test))
```

# LightBGM

```python
import lightgbm as lgb
model = lgb.LGBMClassifier(n_estimators=100)
model.fit(X_train,y_train)

from sklearn.metrics import accuracy_score
pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```

# CatBoost

```python
!pip install catboost
```

```python
from catboost import CatBoostClassifier

model = CatBoostClassifier(n_estimators=100)
model.fit(X_train,y_train)
print(model.score(X_test,y_test))
```

# K-Fold Cross Validation

```python
kfolds_data = new_raw_data.drop(columns=['is_claim'])
kfolds_target = new_raw_data["is_claim"]
```

## XGBoost

### k = 3

```python
from sklearn.model_selection import cross_val_score

XGBoost_k3_result = cross_val_score(XGBClassifier(n_estimators=100), kfolds_data, kfolds_target, cv=3)
print(XGBoost_k3_result)
print(XGBoost_k3_result.mean())
```

### k = 5

```python
XGBoost_k5_result = cross_val_score(XGBClassifier(n_estimators=100), kfolds_data, kfolds_target, cv=5)
print(XGBoost_k5_result)
print(XGBoost_k5_result.mean())
```

### k = 10

```python
XGBoost_k10_result = cross_val_score(XGBClassifier(n_estimators=100), kfolds_data, kfolds_target, cv=10)
print(XGBoost_k10_result)
print(XGBoost_k10_result.mean())
```
